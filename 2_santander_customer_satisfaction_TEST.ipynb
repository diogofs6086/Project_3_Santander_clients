```python
# Packages
import pandas as pd
```

```python
# Datasets
testDF = pd.read_csv('test.csv', index_col = 'ID')
trainDF_9 = pd.read_csv('trainDF_9.csv', index_col = 'ID')
```

```python
# Testing dataset shape
testDF.shape
```

```python
# first five lines of the testing dataset
testDF.head()
```

```python
# Missing values
testDF.isnull().sum().sum()
```

```python
# Take the same columns as trainDF_9
testDF_1 = testDF[trainDF_9.iloc[:,:-1].columns]
```

```python
# Descriptive statistics of the traning dataset
testDF_1.describe()
```

The descriptive statistical parameter of the variables are very differents.

```python
# Package
from sklearn.preprocessing import MinMaxScaler

# Normalization
scaler = MinMaxScaler(feature_range = (0, 1))
testDF_2 = scaler.fit_transform(testDF_1)
testDF_2
```

```python
# Rows and columns nomination
testDF_3 = pd.DataFrame(testDF_2, index = testDF_1.index, columns = testDF_1.columns)
testDF_3
```

```python
# Descriptive statistics of the traning dataset
testDF_3.describe()
```

```python
# Boosting method for training dataset
from sklearn.ensemble import GradientBoostingClassifier

# Dependent variable
Y = trainDF_9.iloc[:,-1]
# Indepent variables
X = trainDF_9.iloc[:,:-1]

# Boosting method
model = GradientBoostingClassifier(loss = 'deviance', learning_rate = 0.05)

# Training the model
model.fit(X, Y)
```

```python
# Prediction for testing dataset
prediction = model.predict(testDF_3)
print('The total of class 0 is', sum(prediction == 0), 'the class 1 is', sum(prediction == 1),'.')
```

```python
# Saving the submission file
pred_submission = pd.Series(prediction, name = 'TARGET', index = testDF_3.index)
pred_submission.to_csv('pred_submission.csv')
```
